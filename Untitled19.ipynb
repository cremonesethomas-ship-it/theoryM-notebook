```python
<!DOCTYPE html>
<html lang="it">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Dashboard Quantistica Avanzata M ≅ S¹ × Γ</title>
    <script src="https://cdn.plot.ly/plotly-2.35.2.min.js"></script>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/PapaParse/5.3.0/papaparse.min.js"></script>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0-alpha1/dist/css/bootstrap.min.css" rel="stylesheet">
    <link rel="stylesheet" href="https://cdnjs.cloud.com/ajax/libs/font-awesome/6.0.0/css/all.min.css">
    <style>
        :root {
            --quantum-primary: #2c3e50;
            --quantum-secondary: #3498db;
            --quantum-accent: #9b59b6;
            --quantum-background: #1a1a2e;
            --quantum-card: #16213e;
            --quantum-text: #e6e6e6;
        }

        body {
            background: linear-gradient(135deg, var(--quantum-background) 0%, #0f3460 100%);
            color: var(--quantum-text);
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            min-height: 100vh;
        }

        .navbar {
            background-color: rgba(26, 26, 46, 0.8);
            backdrop-filter: blur(10px);
            box-shadow: 0 4px 20px rgba(0, 0, 0, 0.25);
        }

        .quantum-card {
            background-color: var(--quantum-card);
            border-radius: 12px;
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.2);
            border: 1px solid rgba(255, 255, 255, 0.1);
            transition: transform 0.3s ease, box-shadow 0.3s ease;
            margin-bottom: 24px;
        }

        .quantum-card:hover {
            transform: translateY(-5px);
            box-shadow: 0 12px 30px rgba(0, 0, 0, 0.3);
        }

        .card-header {
            background: linear-gradient(90deg, var(--quantum-primary) 0%, var(--quantum-accent) 100%);
            border-bottom: 1px solid rgba(255, 255, 255, 0.1);
            font-weight: 600;
            padding: 15px 20px;
        }

        h1, h2, h3 {
            font-weight: 700;
            color: #fff;
        }

        .dashboard-section {
            padding: 30px 0;
        }

        .filter-controls {
            background-color: rgba(255, 255, 255, 0.05);
            border-radius: 8px;
            padding: 15px;
            margin-bottom: 20px;
        }

        .form-control, .form-select {
            background-color: rgba(255, 255, 255, 0.1);
            border: 1px solid rgba(255, 255, 255, 0.2);
            color: var(--quantum-text);
        }

        .form-control:focus, .form-select:focus {
            background-color: rgba(255, 255, 255, 0.15);
            border-color: var(--quantum-secondary);
            box-shadow: 0 0 0 0.25rem rgba(52, 152, 219, 0.25);
            color: var(--quantum-text);
        }

        .table-container {
            overflow-x: auto;
            border-radius: 8px;
        }

        table {
            width: 100%;
            border-collapse: separate;
            border-spacing: 0;
            background-color: var(--quantum-card);
        }

        th {
            background-color: var(--quantum-primary);
            color: white;
            position: sticky;
            top: 0;
            cursor: pointer;
            user-select: none;
        }

        th, td {
            padding: 12px 15px;
            text-align: left;
            border-bottom: 1px solid rgba(255, 255, 255, 0.1);
        }

        tr:hover {
            background-color: rgba(52, 152, 219, 0.1);
        }

        .pagination {
            justify-content: center;
            margin-top: 20px;
        }

        .page-link {
            background-color: var(--quantum-primary);
            border: 1px solid rgba(255, 255, 255, 0.1);
            color: var(--quantum-text);
        }

        .page-link:hover {
            background-color: var(--quantum-secondary);
            border-color: var(--quantum-secondary);
        }

        .page-item.active .page-link {
            background-color: var(--quantum-accent);
            border-color: var(--quantum-accent);
        }

        .chart-container {
            height: 400px;
            margin-bottom: 30px;
        }

        .counts-badge {
            background-color: var(--quantum-accent);
            color: white;
            padding: 3px 8px;
            border-radius: 4px;
            font-size: 0.85em;
            cursor: pointer;
        }

        .counts-modal .modal-content {
            background-color: var(--quantum-card);
            color: var(--quantum-text);
        }

        .glow-text {
            text-shadow: 0 0 10px rgba(155, 89, 182, 0.7);
        }

        .btn-quantum {
            background: linear-gradient(90deg, var(--quantum-secondary) 0%, var(--quantum-accent) 100%);
            border: none;
            color: white;
            transition: all 0.3s ease;
        }

        .btn-quantum:hover {
            transform: translateY(-2px);
            box-shadow: 0 6px 15px rgba(155, 89, 182, 0.4);
        }

        /* Animazioni */
        @keyframes pulse {
            0% { box-shadow: 0 0 0 0 rgba(52, 152, 219, 0.4); }
            70% { box-shadow: 0 0 0 10px rgba(52, 152, 219, 0); }
            100% { box-shadow: 0 0 0 0 rgba(52, 152, 219, 0); }
        }

        .pulse {
            animation: pulse 2s infinite;
        }

        /* Responsive adjustments */
        @media (max-width: 768px) {
            .chart-container {
                height: 300px;
            }

            th, td {
                padding: 8px 10px;
                font-size: 0.9rem;
            }
        }
    </style>
</head>
<body>
    <nav class="navbar navbar-expand-lg navbar-dark sticky-top">
        <div class="container">
            <a class="navbar-brand glow-text" href="#">
                <i class="fas fa-atom me-2"></i>Dashboard Quantistica M ≅ S¹ × Γ
            </a>
            <button class="navbar-toggler" type="button" data-bs-toggle="collapse" data-bs-target="#navbarContent">
                <span class="navbar-toggler-icon"></span>
            </button>
            <div class="collapse navbar-collapse" id="navbarContent">
                <ul class="navbar-nav ms-auto">
                    <li class="nav-item">
                        <a class="nav-link active" href="#dati"><i class="fas fa-table me-1"></i> Dati</a>
                    </li>
                    <li class="nav-item">
                        <a class="nav-link" href="#grafici"><i class="fas fa-chart-line me-1"></i> Grafici</a>
                    </li>
                    <li class="nav-item">
                        <a class="nav-link" href="#informazioni"><i class="fas fa-info-circle me-1"></i> Informazioni</a>
                    </li>
                </ul>
            </div>
        </div>
    </nav>

    <div class="container dashboard-section">
        <section id="dati">
            <div class="quantum-card">
                <div class="card-header">
                    <h2 class="mb-0"><i class="fas fa-table me-2"></i>Tabella dei risultati (ordinabile e filtrabile)</h2>
                </div>
                <div class="card-body">
                    <div class="filter-controls mb-4">
                        <div class="row g-3">
                            <div class="col-md-3">
                                <label for="searchInput" class="form-label">Cerca</label>
                                <input type="text" class="form-control" id="searchInput" placeholder="Cerca...">
                            </div>
                            <div class="col-md-2">
                                <label for="qubitsFilter" class="form-label">Qubits</label>
                                <select class="form-select" id="qubitsFilter">
                                    <option value="">Tutti</option>
                                    <option value="2">2</option>
                                    <option value="3">3</option>
                                    <option value="4">4</option>
                                    <option value="5">5</option>
                                    <option value="6">6</option>
                                </select>
                            </div>
                            <div class="col-md-3">
                                <label for="entTypeFilter" class="form-label">Tipo di Entanglement</label>
                                <select class="form-select" id="entTypeFilter">
                                    <option value="">Tutti</option>
                                    <option value="linear">Linear</option>
                                    <option value="grid">Grid</option>
                                    <option value="all_to_all">All-to-all</option>
                                </select>
                            </div>
                            <div class="col-md-2">
                                <label for="rowsPerPage" class="form-label">Righe per pagina</label>
                                <select class="form-select" id="rowsPerPage">
                                    <option value="10">10</option>
                                    <option value="25">25</option>
                                    <option value="50">50</option>
                                    <option value="100">100</option>
                                </select>
                            </div>
                            <div class="col-md-2 d-flex align-items-end">
                                <button id="resetFilters" class="btn btn-outline-light w-100">
                                    <i class="fas fa-sync-alt me-1"></i> Reset
                                </button>
                            </div>
                        </div>
                    </div>

                    <div class="table-container">
                        <table class="table table-hover" id="dataTable">
                            <thead>
                                <tr>
                                    <th data-sort="index"># <i class="fas fa-sort"></i></th>
                                    <th data-sort="qubits">Qubits <i class="fas fa-sort"></i></th>
                                    <th data-sort="ent_type">Tipo Ent. <i class="fas fa-sort"></i></th>
                                    <th data-sort="theta">Theta <i class="fas fa-sort"></i></th>
                                    <th data-sort="entropy">Entropia <i class="fas fa-sort"></i></th>
                                    <th data-sort="max_prob">Max Prob <i class="fas fa-sort"></i></th>
                                    <th data-sort="states"># Stati <i class="fas fa-sort"></i></th>
                                    <th>Counts</th>
                                </tr>
                            </thead>
                            <tbody id="tableBody">
                                <!-- I dati verranno popolati via JavaScript -->
                            </tbody>
                        </table>
                    </div>

                    <div class="d-flex justify-content-between align-items-center mt-3">
                        <div id="tableInfo" class="text-muted">Mostrando 0 di 0 elementi</div>
                        <nav>
                            <ul class="pagination" id="pagination">
                                <li class="page-item disabled">
                                    <a class="page-link" href="#" id="prevPage">Precedente</a>
                                </li>
                                <li class="page-item active">
                                    <a class="page-link" href="#">1</a>
                                </li>
                                <li class="page-item disabled">
                                    <a class="page-link" href="#" id="nextPage">Successiva</a>
                                </li>
                            </ul>
                        </nav>
                    </div>
                </div>
            </div>
        </section>

        <section id="grafici" class="mt-5">
            <div class="quantum-card">
                <div class="card-header">
                    <h2 class="mb-0"><i class="fas fa-chart-line me-2"></i>Grafici interattivi</h2>
                </div>
                <div class="card-body">
                    <div class="filter-controls mb-4">
                        <div class="row g-3">
                            <div class="col-md-4">
                                <label for="chartQubits" class="form-label">Qubits da visualizzare</label>
                                <select class="form-select" id="chartQubits" multiple>
                                    <option value="2" selected>2</option>
                                    <option value="3" selected>3</option>
                                    <option value="4" selected>4</option>
                                    <option value="5" selected>5</option>
                                    <option value="6" selected>6</option>
                                </select>
                            </div>
                            <div class="col-md-4">
                                <label for="chartEntType" class="form-label">Tipo di Entanglement</label>
                                <select class="form-select" id="chartEntType" multiple>
                                    <option value="linear" selected>Linear</option>
                                    <option value="grid" selected>Grid</option>
                                    <option value="all_to_all" selected>All-to-all</option>
                                </select>
                            </div>
                            <div class="col-md-4 d-flex align-items-end">
                                <button id="updateCharts" class="btn btn-quantum w-100">
                                    <i class="fas fa-sync me-1"></i> Aggiorna Grafici
                                </button>
                            </div>
                        </div>
                    </div>

                    <div class="row">
                        <div class="col-lg-6">
                            <div class="quantum-card">
                                <div class="card-header">
                                    <h3 class="mb-0">Entropia di Shannon vs θ</h3>
                                </div>
                                <div class="card-body">
                                    <div id="entropyChart" class="chart-container"></div>
                                </div>
                            </div>
                        </div>
                        <div class="col-lg-6">
                            <div class="quantum-card">
                                <div class="card-header">
                                    <h3 class="mb-0">Probabilità Massima vs θ</h3>
                                </div>
                                <div class="card-body">
                                    <div id="probabilityChart" class="chart-container"></div>
                                </div>
                            </div>
                        </div>
                    </div>

                    <div class="row mt-4">
                        <div class="col-12">
                            <div class="quantum-card">
                                <div class="card-header">
                                    <h3 class="mb-0">Distribuzioni di Counts</h3>
                                </div>
                                <div class="card-body">
                                    <div class="row mb-4">
                                        <div class="col-md-4">
                                            <select class="form-select" id="countsSelector">
                                                <option value="0">2q linear θ=0.79</option>
                                                <option value="1">2q linear θ=1.05</option>
                                                <option value="2">2q linear θ=1.57</option>
                                                <option value="3">2q grid θ=0.79</option>
                                                <option value="4">2q grid θ=1.05</option>
                                                <option value="5">2q grid θ=1.57</option>
                                                <option value="6">2q all_to_all θ=0.79</option>
                                                <option value="7">2q all_to_all θ=1.05</option>
                                                <option value="8">2q all_to_all θ=1.57</option>
                                                <option value="9">3q linear θ=0.79</option>
                                                <option value="10">3q linear θ=1.05</option>
                                                <option value="11">3q linear θ=1.57</option>
                                                <option value="12">3q grid θ=0.79</option>
                                            </select>
                                        </div>
                                    </div>
                                    <div id="countsChart" class="chart-container"></div>
                                </div>
                            </div>
                        </div>
                    </div>
                </div>
            </div>
        </section>

        <section id="informazioni" class="mt-5">
            <div class="quantum-card">
                <div class="card-header">
                    <h2 class="mb-0"><i class="fas fa-info-circle me-2"></i>Informazioni</h2>
                </div>
                <div class="card-body">
                    <div class="row">
                        <div class="col-md-6">
                            <h4>Descrizione del Progetto</h4>
                            <p>Questa dashboard visualizza i risultati di esperimenti quantistici che esplorano diversi tipi di entanglement (lineare, griglia e all-to-all) con vari numeri di qubit (da 2 a 6) e angoli theta differenti.</p>
                            <p>I dati mostrano l'entropia di Shannon, la probabilità massima e le distribuzioni di stati misurati.</p>
                        </div>
                        <div class="col-md-6">
                            <h4>Parametri</h4>
                            <ul>
                                <li><strong>Qubits</strong>: Numero di qubit nel sistema (2-6)</li>
                                <li><strong>Tipo di Entanglement</strong>: Configurazione dei collegamenti tra qubit</li>
                                <li><strong>Theta (θ)</strong>: Angolo di rotazione applicato</li>
                                <li><strong>Entropia</strong>: Entropia di Shannon della distribuzione di probabilità</li>
                                <li><strong>Max Prob</strong>: Probabilità massima tra tutti gli stati</li>
                                <li><strong>Counts</strong>: Distribuzione degli stati misurati</li>
                            </ul>
                        </div>
                    </div>
                </div>
            </div>
        </section>
    </div>

    <!-- Modal per visualizzare i counts -->
    <div class="modal fade counts-modal" id="countsModal" tabindex="-1">
        <div class="modal-dialog modal-lg">
            <div class="modal-content">
                <div class="modal-header">
                    <h5 class="modal-title">Dettagli Counts</h5>
                    <button type="button" class="btn-close btn-close-white" data-bs-dismiss="modal"></button>
                </div>
                <div class="modal-body">
                    <pre id="countsDetails" class="p-3 bg-dark text-light rounded"></pre>
                </div>
                <div class="modal-footer">
                    <button type="button" class="btn btn-secondary" data-bs-dismiss="modal">Chiudi</button>
                </div>
            </div>
        </div>
    </div>

    <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0-alpha1/dist/js/bootstrap.bundle.min.js"></script>
    <script>
        // Dati dalla tabella originale
        const tableData = [
            {qubits: 2, ent_type: 'linear', theta: 0.785398, entropy: 0.592042, max_prob: 0.8570, states: 2, counts: {'11': 286, '00': 1714}},
            {qubits: 2, ent_type: 'linear', theta: 1.047198, entropy: 0.764706, max_prob: 0.7775, states: 2, counts: {'11': 445, '00': 1555}},
            {qubits: 2, ent_type: 'linear', theta: 1.570796, entropy: 0.999512, max_prob: 0.5130, states: 2, counts: {'00': 1026, '11': 974}},
            {qubits: 2, ent_type: 'grid', theta: 0.785398, entropy: 0.589453, max_prob: 0.8580, states: 2, counts: {'11': 284, '00': 1716}},
            {qubits: 2, ent_type: 'grid', theta: 1.047198, entropy: 0.828252, max_prob: 0.7390, states: 2, counts: {'00': 1478, '11': 522}},
            {qubits: 2, ent_type: 'grid', theta: 1.570796, entropy: 0.999974, max_prob: 0.5030, states: 2, counts: {'11': 1006, '00': 994}},
            {qubits: 2, ent_type: 'all_to_all', theta: 0.785398, entropy: 0.588154, max_prob: 0.8585, states: 2, counts: {'00': 1717, '10': 283}},
            {qubits: 2, ent_type: 'all_to_all', theta: 1.047198, entropy: 0.808093, max_prob: 0.7520, states: 2, counts: {'10': 496, '00': 1504}},
            {qubits: 2, ent_type: 'all_to_all', theta: 1.570796, entropy: 0.999474, max_prob: 0.5135, states: 2, counts: {'00': 973, '10': 1027}},
            {qubits: 3, ent_type: 'linear', theta: 0.785398, entropy: 0.608588, max_prob: 0.8505, states: 2, counts: {'111': 299, '000': 1701}},
            {qubits: 3, ent_type: 'linear', theta: 1.047198, entropy: 0.846442, max_prob: 0.7265, states: 2, counts: {'000': 1453, '111': 547}},
            {qubits: 3, ent_type: 'linear', theta: 1.570796, entropy: 0.999618, max_prob: 0.5115, states: 2, counts: {'000': 1023, '111': 977}},
            {qubits: 3, ent_type: 'grid', theta: 0.785398, entropy: 0.608588, max_prob: 0.8505, states: 2, counts: {'111': 299, '000': 1701}},
            {qubits: 3, ent_type: 'grid', theta: 1.047198, entropy: 0.822950, max_prob: 0.7425, states: 2, counts: {'000': 1485, '111': 515}},
            {qubits: 3, ent_type: 'grid', theta: 1.570796, entropy: 0.999166, max_prob: 0.5170, states: 2, counts: {'111': 966, '000': 1034}},
            {qubits: 3, ent_type: 'all_to_all', theta: 0.785398, entropy: 0.620988, max_prob: 0.8455, states: 2, counts: {'010': 309, '000': 1691}},
            {qubits: 3, ent_type: 'all_to_all', theta: 1.047198, entropy: 0.813647, max_prob: 0.7485, states: 2, counts: {'010': 503, '000': 1497}},
            {qubits: 3, ent_type: 'all_to_all', theta: 1.570796, entropy: 0.999994, max_prob: 0.5015, states: 2, counts: {'000': 1003, '010': 997}},
            {qubits: 4, ent_type: 'linear', theta: 0.785398, entropy: 0.551928, max_prob: 0.8720, states: 2, counts: {'1111': 256, '0000': 1744}},
            {qubits: 4, ent_type: 'linear', theta: 1.047198, entropy: 0.802444, max_prob: 0.7555, states: 2, counts: {'0000': 1511, '1111': 489}},
            {qubits: 4, ent_type: 'linear', theta: 1.570796, entropy: 0.999974, max_prob: 0.5030, states: 2, counts: {'1111': 1006, '0000': 994}},
            {qubits: 4, ent_type: 'grid', theta: 0.785398, entropy: 0.597185, max_prob: 0.8550, states: 2, counts: {'1111': 290, '0000': 1710}},
            {qubits: 4, ent_type: 'grid', theta: 1.047198, entropy: 0.796699, max_prob: 0.7590, states: 2, counts: {'1111': 482, '0000': 1518}},
            {qubits: 4, ent_type: 'grid', theta: 1.570796, entropy: 0.999913, max_prob: 0.5055, states: 2, counts: {'0000': 1011, '1111': 989}},
            {qubits: 4, ent_type: 'all_to_all', theta: 0.785398, entropy: 0.612337, max_prob: 0.8490, states: 2, counts: {'0010': 302, '0000': 1698}},
            {qubits: 4, ent_type: 'all_to_all', theta: 1.047198, entropy: 0.781481, max_prob: 0.7680, states: 2, counts: {'0010': 464, '0000': 1536}},
            {qubits: 4, ent_type: 'all_to_all', theta: 1.570796, entropy: 0.999988, max_prob: 0.5020, states: 2, counts: {'0010': 996, '0000': 1004}},
            {qubits: 5, ent_type: 'linear', theta: 0.785398, entropy: 0.598464, max_prob: 0.8545, states: 2, counts: {'11111': 291, '00000': 1709}},
            {qubits: 5, ent_type: 'linear', theta: 1.047198, entropy: 0.827500, max_prob: 0.7395, states: 2, counts: {'00000': 1479, '11111': 521}},
            {qubits: 5, ent_type: 'linear', theta: 1.570796, entropy: 0.999859, max_prob: 0.5070, states: 2, counts: {'00000': 1014, '11111': 986}},
            {qubits: 5, ent_type: 'grid', theta: 0.785398, entropy: 0.580297, max_prob: 0.8615, states: 2, counts: {'11111': 277, '00000': 1723}},
            {qubits: 5, ent_type: 'grid', theta: 1.047198, entropy: 0.819107, max_prob: 0.7450, states: 2, counts: {'11111': 510, '00000': 1490}},
            {qubits: 5, ent_type: 'grid', theta: 1.570796, entropy: 0.999393, max_prob: 0.5145, states: 2, counts: {'00000': 971, '11111': 1029}},
            {qubits: 5, ent_type: 'all_to_all', theta: 0.785398, entropy: 0.641438, max_prob: 0.8370, states: 2, counts: {'00010': 326, '00000': 1674}},
            {qubits: 5, ent_type: 'all_to_all', theta: 1.047198, entropy: 0.845029, max_prob: 0.7275, states: 2, counts: {'00010': 545, '00000': 1455}},
            {qubits: 5, ent_type: 'all_to_all', theta: 1.570796, entropy: 0.999878, max_prob: 0.5065, states: 2, counts: {'00000': 987, '00010': 1013}},
            {qubits: 6, ent_type: 'linear', theta: 0.785398, entropy: 0.612337, max_prob: 0.8490, states: 2, counts: {'111111': 302, '000000': 1698}},
            {qubits: 6, ent_type: 'linear', theta: 1.047198, entropy: 0.822950, max_prob: 0.7425, states: 2, counts: {'111111': 515, '000000': 1485}},
            {qubits: 6, ent_type: 'linear', theta: 1.570796, entropy: 0.999393, max_prob: 0.5145, states: 2, counts: {'111111': 1029, '000000': 971}},
            {qubits: 6, ent_type: 'grid', theta: 0.785398, entropy: 0.618530, max_prob: 0.8465, states: 2, counts: {'111111': 307, '000000': 1693}},
            {qubits: 6, ent_type: 'grid', theta: 1.047198, entropy: 0.800812, max_prob: 0.7565, states: 2, counts: {'111111': 487, '000000': 1513}},
            {qubits: 6, ent_type: 'grid', theta: 1.570796, entropy: 0.999954, max_prob: 0.5040, states: 2, counts: {'000000': 992, '111111': 1008}},
            {qubits: 6, ent_type: 'all_to_all', theta: 0.785398, entropy: 0.617297, max_prob: 0.8470, states: 2, counts: {'000010': 306, '000000': 1694}},
            {qubits: 6, ent_type: 'all_to_all', theta: 1.047198, entropy: 0.778011, max_prob: 0.7700, states: 2, counts: {'000010': 460, '000000': 1540}},
            {qubits: 6, ent_type: 'all_to_all', theta: 1.570796, entropy: 0.999792, max_prob: 0.5085, states: 2, counts: {'000000': 983, '000010': 1017}}
        ];

        // Aggiungi un indice a ciascun elemento
        tableData.forEach((item, index) => {
            item.index = index;
        });

        // Variabili globali per lo stato della tabella
        let currentPage = 1;
        let rowsPerPage = 10;
        let currentSort = { column: 'index', direction: 'asc' };
        let filteredData = [...tableData];

        // Inizializzazione
        document.addEventListener('DOMContentLoaded', function() {
            initializeTable();
            initializeCharts();
            setupEventListeners();
        });

        // Inizializza la tabella
        function initializeTable() {
            filteredData = [...tableData];
            renderTable();
            updatePagination();
        }

        // Imposta i gestori degli eventi
        function setupEventListeners() {
            // Filtri
            document.getElementById('searchInput').addEventListener('input', filterTable);
            document.getElementById('qubitsFilter').addEventListener('change', filterTable);
            document.getElementById('entTypeFilter').addEventListener('change', filterTable);
            document.getElementById('rowsPerPage').addEventListener('change', function() {
                rowsPerPage = parseInt(this.value);
                currentPage = 1;
                renderTable();
                updatePagination();
            });

            document.getElementById('resetFilters').addEventListener('click', function() {
                document.getElementById('searchInput').value = '';
                document.getElementById('qubitsFilter').value = '';
                document.getElementById('entTypeFilter').value = '';
                rowsPerPage = 10;
                document.getElementById('rowsPerPage').value = '10';
                currentSort = { column: 'index', direction: 'asc' };
                initializeTable();
            });

            // Ordinamento
            document.querySelectorAll('th[data-sort]').forEach(header => {
                header.addEventListener('click', () => {
                    const column = header.getAttribute('data-sort');

                    if (currentSort.column === column) {
                        currentSort.direction = currentSort.direction === 'asc' ? 'desc' : 'asc';
                    } else {
                        currentSort.column = column;
                        currentSort.direction = 'asc';
                    }

                    sortTable(column, currentSort.direction);
                    renderTable();
                });
            });

            // Paginazione
            document.getElementById('prevPage').addEventListener('click', function(e) {
                e.preventDefault();
                if (currentPage > 1) {
                    currentPage--;
                    renderTable();
                    updatePagination();
                }
            });

            document.getElementById('nextPage').addEventListener('click', function(e) {
                e.preventDefault();
                const totalPages = Math.ceil(filteredData.length / rowsPerPage);
                if (currentPage < totalPages) {
                    currentPage++;
                    renderTable();
                    updatePagination();
                }
            });

            // Grafici
            document.getElementById('updateCharts').addEventListener('click', updateCharts);
            document.getElementById('countsSelector').addEventListener('change', updateCountsChart);

            // Navigazione
            document.querySelectorAll('.nav-link').forEach(link => {
                link.addEventListener('click', function(e) {
                    e.preventDefault();
                    const targetId = this.getAttribute('href').substring(1);
                    document.getElementById(targetId).scrollIntoView({ behavior: 'smooth' });

                    // Aggiorna classe attiva
                    document.querySelectorAll('.nav-link').forEach(navLink => {
                        navLink.classList.remove('active');
                    });
                    this.classList.add('active');
                });
            });
        }

        // Filtra la tabella in base ai criteri
        function filterTable() {
            const searchText = document.getElementById('searchInput').value.toLowerCase();
            const qubitsFilter = document.getElementById('qubitsFilter').value;
            const entTypeFilter = document.getElementById('entTypeFilter').value;

            filteredData = tableData.filter(item => {
                return (
                    (searchText === '' ||
                     Object.values(item).some(val =>
                         String(val).toLowerCase().includes(searchText)
                     )) &&
                    (qubitsFilter === '' || item.qubits == qubitsFilter) &&
                    (entTypeFilter === '' || item.ent_type === entTypeFilter)
                );
            });

            currentPage = 1;
            renderTable();
            updatePagination();
        }

        // Ordina la tabella
        function sortTable(column, direction) {
            filteredData.sort((a, b) => {
                let valueA = a[column];
                let valueB = b[column];

                if (typeof valueA === 'string') {
                    valueA = valueA.toLowerCase();
                    valueB = valueB.toLowerCase();
                }

                if (valueA < valueB) return direction === 'asc' ? -1 : 1;
                if (valueA > valueB) return direction === 'asc' ? 1 : -1;
                return 0;
            });
        }

        // Renderizza la tabella
        function renderTable() {
            const tableBody = document.getElementById('tableBody');
            tableBody.innerHTML = '';

            const startIndex = (currentPage - 1) * rowsPerPage;
            const endIndex = Math.min(startIndex + rowsPerPage, filteredData.length);

            for (let i = startIndex; i < endIndex; i++) {
                const item = filteredData[i];
                const row = document.createElement('tr');

                row.innerHTML = `
                    <td>${item.index}</td>
                    <td>${item.qubits}</td>
                    <td>${item.ent_type}</td>
                    <td>${item.theta.toFixed(6)}</td>
                    <td>${item.entropy.toFixed(6)}</td>
                    <td>${item.max_prob.toFixed(4)}</td>
                    <td>${item.states}</td>
                    <td>
                        <span class="counts-badge" data-index="${i}">
                            <i class="fas fa-chart-bar me-1"></i>Visualizza
                        </span>
                    </td>
                `;

                tableBody.appendChild(row);
            }

            // Aggiungi gestore eventi per i badge dei counts
            document.querySelectorAll('.counts-badge').forEach(badge => {
                badge.addEventListener('click', function() {
                    const index = parseInt(this.getAttribute('data-index'));
                    showCountsDetails(filteredData[index].counts);
                });
            });

            // Aggiorna informazioni tabella
            document.getElementById('tableInfo').textContent =
                `Mostrando ${startIndex + 1} a ${endIndex} di ${filteredData.length} elementi`;
        }

        // Aggiorna la paginazione
        function updatePagination() {
            const pagination = document.getElementById('pagination');
            const totalPages = Math.ceil(filteredData.length / rowsPerPage);

            // Pulsante Precedente
            const prevPage = document.getElementById('prevPage');
            prevPage.parentElement.classList.toggle('disabled', currentPage === 1);

            // Pulsante Successiva
            const nextPage = document.getElementById('nextPage');
            nextPage.parentElement.classList.toggle('disabled', currentPage === totalPages);

            // Numeri di pagina
            const pageNumbers = pagination.querySelectorAll('.page-number');
            pageNumbers.forEach(page => page.remove());

            // Calcola range di pagine da mostrare
            let startPage = Math.max(1, currentPage - 2);
            let endPage = Math.min(totalPages, startPage + 4);

            if (endPage - startPage < 4) {
                startPage = Math.max(1, endPage - 4);
            }

            // Aggiungi numeri di pagina
            const prevPageItem = pagination.querySelector('#prevPage').parentElement;

            for (let i = startPage; i <= endPage; i++) {
                const pageItem = document.createElement('li');
                pageItem.className = `page-item page-number ${i === currentPage ? 'active' : ''}`;

                const pageLink = document.createElement('a');
                pageLink.className = 'page-link';
                pageLink.href = '#';
                pageLink.textContent = i;

                pageLink.addEventListener('click', function(e) {
                    e.preventDefault();
                    currentPage = i;
                    renderTable();
                    updatePagination();
                });

                pageItem.appendChild(pageLink);
                pagination.insertBefore(pageItem, nextPage.parentElement);
            }
        }

        // Mostra i dettagli dei counts in un modal
        function showCountsDetails(counts) {
            const countsDetails = document.getElementById('countsDetails');
            countsDetails.textContent = JSON.stringify(counts, null, 2);

            const countsModal = new bootstrap.Modal(document.getElementById('countsModal'));
            countsModal.show();
        }

        // Inizializza i grafici
        function initializeCharts() {
            updateEntropyChart();
            updateProbabilityChart();
            updateCountsChart();
        }

        // Aggiorna i grafici in base ai filtri
        function updateCharts() {
            updateEntropyChart();
            updateProbabilityChart();
        }

        // Grafico dell'entropia
        function updateEntropyChart() {
            const selectedQubits = Array.from(document.getElementById('chartQubits').selectedOptions)
                .map(option => parseInt(option.value));
            const selectedEntTypes = Array.from(document.getElementById('chartEntType').selectedOptions)
                .map(option => option.value);

            const filteredData = tableData.filter(item =>
                selectedQubits.includes(item.qubits) && selectedEntTypes.includes(item.ent_type)
            );

            // Raggruppa per tipo di entanglement e numero di qubit
            const traces = [];
            const colors = ['#636efa', '#EF553B', '#00cc96', '#ab63fa', '#FFA15A', '#19d3f3'];
            const dashStyles = ['solid', 'dot', 'dash', 'longdash', 'dashdot'];

            let colorIndex = 0;
            let dashIndex = 0;

            for (const entType of selectedEntTypes) {
                for (const qubit of selectedQubits) {
                    const dataForTrace = filteredData.filter(item =>
                        item.ent_type === entType && item.qubits === qubit
                    );

                    if (dataForTrace.length > 0) {
                        const trace = {
                            x: dataForTrace.map(item => item.theta),
                            y: dataForTrace.map(item => item.entropy),
                            mode: 'lines+markers',
                            name: `${entType}, ${qubit}`,
                            line: {
                                color: colors[colorIndex % colors.length],
                                dash: dashStyles[dashIndex % dashStyles.length]
                            },
                            marker: {
                                size: 8
                            }
                        };

                        traces.push(trace);
                        colorIndex++;
                        dashIndex++;
                    }
                }
            }

            const layout = {
                title: 'Entropia di Shannon vs θ',
                xaxis: {
                    title: 'θ [rad]',
                    gridcolor: 'rgba(255, 255, 255, 0.1)',
                    zerolinecolor: 'rgba(255, 255, 255, 0.3)'
                },
                yaxis: {
                    title: 'Entropia [bit]',
                    gridcolor: 'rgba(255, 255, 255, 0.1)',
                    zerolinecolor: 'rgba(255, 255, 255, 0.3)'
                },
                plot_bgcolor: 'rgba(0, 0, 0, 0)',
                paper_bgcolor: 'rgba(0, 0, 0, 0)',
                font: {
                    color: '#e6e6e6'
                },
                legend: {
                    x: 0,
                    y: 1.1,
                    orientation: 'h'
                },
                margin: {
                    l: 60,
                    r: 40,
                    b: 50,
                    t: 80,
                    pad: 4
                }
            };

            Plotly.newPlot('entropyChart', traces, layout, {responsive: true});
        }

        // Grafico della probabilità
        function updateProbabilityChart() {
            const selectedQubits = Array.from(document.getElementById('chartQubits').selectedOptions)
                .map(option => parseInt(option.value));
            const selectedEntTypes = Array.from(document.getElementById('chartEntType').selectedOptions)
                .map(option => option.value);

            const filteredData = tableData.filter(item =>
                selectedQubits.includes(item.qubits) && selectedEntTypes.includes(item.ent_type)
            );

            // Raggruppa per tipo di entanglement e numero di qubit
            const traces = [];
            const colors = ['#636efa', '#EF553B', '#00cc96', '#ab63fa', '#FFA15A', '#19d3f3'];
            const dashStyles = ['solid', 'dot', 'dash', 'longdash', 'dashdot'];

            let colorIndex = 0;
            let dashIndex = 0;

            for (const entType of selectedEntTypes) {
                for (const qubit of selectedQubits) {
                    const dataForTrace = filteredData.filter(item =>
                        item.ent_type === entType && item.qubits === qubit
                    );

                    if (dataForTrace.length > 0) {
                        const trace = {
                            x: dataForTrace.map(item => item.theta),
                            y: dataForTrace.map(item => item.max_prob),
                            mode: 'lines+markers',
                            name: `${entType}, ${qubit}`,
                            line: {
                                color: colors[colorIndex % colors.length],
                                dash: dashStyles[dashIndex % dashStyles.length]
                            },
                            marker: {
                                size: 8
                            }
                        };

                        traces.push(trace);
                        colorIndex++;
                        dashIndex++;
                    }
                }
            }

            const layout = {
                title: 'Probabilità Massima vs θ',
                xaxis: {
                    title: 'θ [rad]',
                    gridcolor: 'rgba(255, 255, 255, 0.1)',
                    zerolinecolor: 'rgba(255, 255, 255, 0.3)'
                },
                yaxis: {
                    title: 'Max Prob',
                    gridcolor: 'rgba(255, 255, 255, 0.1)',
                    zerolinecolor: 'rgba(255, 255, 255, 0.3)'
                },
                plot_bgcolor: 'rgba(0, 0, 0, 0)',
                paper_bgcolor: 'rgba(0, 0, 0, 0)',
                font: {
                    color: '#e6e6e6'
                },
                legend: {
                    x: 0,
                    y: 1.1,
                    orientation: 'h'
                },
                margin: {
                    l: 60,
                    r: 40,
                    b: 50,
                    t: 80,
                    pad: 4
                }
            };

            Plotly.newPlot('probabilityChart', traces, layout, {responsive: true});
        }

        // Grafico dei counts
        function updateCountsChart() {
            const selectedIndex = parseInt(document.getElementById('countsSelector').value);
            const selectedData = tableData[selectedIndex];

            if (!selectedData) return;

            const counts = selectedData.counts;
            const states = Object.keys(counts);
            const values = Object.values(counts);

            const trace = {
                x: states,
                y: values,
                type: 'bar',
                marker: {
                    color: '#9b59b6',
                    line: {
                        color: '#8e44ad',
                        width: 1.5
                    }
                }
            };

            const layout = {
                title: `Distribuzione Counts: ${selectedData.qubits}q ${selectedData.ent_type} θ=${selectedData.theta.toFixed(2)}`,
                xaxis: {
                    title: 'Stato',
                    gridcolor: 'rgba(255, 255, 255, 0.1)'
                },
                yaxis: {
                    title: 'Counts',
                    gridcolor: 'rgba(255, 255, 255, 0.1)'
                },
                plot_bgcolor: 'rgba(0, 0, 0, 0)',
                paper_bgcolor: 'rgba(0, 0, 0, 0)',
                font: {
                    color: '#e6e6e6'
                },
                margin: {
                    l: 60,
                    r: 40,
                    b: 60,
                    t: 80,
                    pad: 4
                }
            };

            Plotly.newPlot('countsChart', [trace], layout, {responsive: true});
        }
    </script>
</body>
</html>
```
